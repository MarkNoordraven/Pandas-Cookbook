```python
import matplotlib.pyplot as plt
```

```python
x = [-3,5,7]
y = [10,2,5]
```

```python
plt.figure(figsize = (15,3))
```

```python
plt.plot(x,y)
```

```python
plt.xlim(0,10)
```

```python
plt.ylim(-3,8)
```

```python
plt.xlabel('X Axis')
```

```python
plt.ylabel('Y axis')
```

```python
plt.title('Line Plot')
```

```python
plt.suptitle('Figure Title',
             size = 30,
             y = 1.03)
```

```python
figure, ax = plt.subplots(figsize = (15, 3))
```

```python
ax.plot(x, y)
```

```python
ax.set_xlim(0, 10)
```

```python
ax.set_ylim(-3, 8)
```

```python
ax.set_xlabel('X axis')
```

```python
ax.set_ylabel('Y axis')
```

```python
ax.set_title('Line Plot')
```

```python
figure.suptitle('Figure Title',
                size = 20,
                y = 1.03)
```
