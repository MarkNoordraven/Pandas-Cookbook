```python
slb = pd.read_csv('data/slb_stock.csv', index_col='Date', parse_dates=['Date'])
slb.head()
```

```python
slb_close = slb['Close']
slb_close
```

```python
slb_summary = slb_close.describe(percentiles=[0.1, 0.9])
slb_summary
```

```python
slb_close.describe()
```

```python
slb_summary.loc['min']
```

```python
list(slb['Open'])
```

```python
slb.head()
```

```python
upper_10 = slb_summary.loc['90%']
lower_10 = slb_summary.loc['10%']
criteria = (slb_close < lower_10) | (slb_close > upper_10)
slb_top_bottom_10 = slb_close[criteria]
slb_top_bottom_10
```

```python
import matplotlib

slb_close.plot(color='black', figsize=(12,6))
slb_top_bottom_10.plot(marker='o', style=' ', ms=4, color='lightgray')


xmin = criteria.index[0]
xmax = criteria.index[-1]
matplotlib.pyplot.hlines(y=[lower_10, upper_10], xmin=xmin, xmax=xmax, color='lightgray');
```

```python
slb_close.plot(color='black', figsize=(12,6))
matplotlib.pyplot.hlines(y=[lower_10, upper_10], xmin=xmin, xmax=xmax, color='lightgray')
matplotlib.pyplot.fill_between(x=criteria.index, y1=lower_10, y2=slb_close.values, color='black')
matplotlib.pyplot.fill_between(x=criteria.index, y1=lower_10, y2=slb_close.values, where=slb_close<lower_10, color='lightgray')
matplotlib.pyplot.fill_between(x=criteria.index, y1=upper_10, y2=slb_close.values, where=slb_close>upper_10, color='lightgray')
```
