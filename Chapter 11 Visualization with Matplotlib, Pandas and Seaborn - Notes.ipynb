```python
import pandas as pd
movie = pd.read_csv('data/movie.csv')
movie.head()
```

```python
med_budget = movie.groupby('title_year')['budget'].median() / 1e6
med_budget
```

```python
med_budget_roll = med_budget.rolling(5,
                                     min_periods = 1)\
                            .mean()
med_budget_roll
```

```python
years = med_budget_roll.index.values
years[-5:]
```

```python
budget = med_budget_roll.values
budget[-5:]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (14,4),
                       linewidth = 5,
                       edgecolor = '.5')
ax.plot(years,
        budget,
        linestyle = '--',
        linewidth = 3,
        color = '0.2',
        label = 'All Movies');
```

```python
text_kwargs = dict(fontsize=20,
                   family = 'cursive')
ax.set_title('Median Movie Budget',
             **text_kwargs)
ax.set_ylabel('Millions of Dollars',
              **text_kwargs)
fig
```

```python
movie_count = movie.groupby('title_year')['budget'].count()
movie_count.tail()
```

```python
ct = movie_count.values
ct_norm = ct / ct.max() * budget.max()
ct_norm
```

```python
fifth_year = (years % 5 == 0) & (years >= 1970)
years_5 = years[fifth_year]
years
```

```python
ct_5 = ct[fifth_year]
ct_5
```

```python
ct_norm_5 = ct_norm[fifth_year]
ct_norm_5
```

```python
ax.bar(years_5,
       ct_norm_5,
       3,
       facecolor = '0.5',
       alpha = 0.3,
       label = 'Movies per Year')
ax.set_xlim
fig
```

```python
for x,y,v in zip(years_5, ct_norm_5, ct_5):
    ax.text(x, y+0.5, str(v), ha = 'center')
ax.legend()
fig
```

```python
top10 = movie.sort_values('budget',
                          ascending = False)\
             .groupby('title_year')['budget']\
             .apply(lambda x: x.iloc[:10].median() / 1e6)
top10
```

```python
 top10_roll = top10.rolling(5, min_periods=1).mean()
```

```python
fig2, ax_array = plt.subplots(2,
                              1,
                              figsize = (14,8),
                              sharex = True)
ax1 = ax_array[0]
ax2 = ax_array[1]
```

```python
ax1.plot(years,
         budget,
         linestyle = '--',
         linewidth = 3,
         color ='0.2',
         label = 'All Movies')
```

```python
ax1.bar(years_5,
        ct_norm_5,
        3,
        facecolor = '0.5',
        alpha= 0.3,
        label = 'Movies per Year')
ax1.legend(loc = 'upper left')
ax1.set_xlim(1968, 2018)

for x, y, v in zip(years_5, ct_norm_5, ct_5):
        ax1.text(x, y + .5, str(v), ha='center')

ax2.plot(years,
         top10_roll.values,
         color = '0.2',
         label = 'Top 10 Movies')
ax2.legend(loc = 'upper left')

fig2.tight_layout()
fig2
```

```python
import numpy as np

cols = ['budget', 'title_year', 'imdb_score', 'movie_title']
m =movie[cols].dropna()
m['budget2'] = m['budget'] / 1e6
np.random.seed(0)
movie_samp = m.query('title_year >= 2000').sample(100)


fig, ax = plt.subplots(figsize = (14,6))
ax.scatter(x = 'title_year',
           y = 'imdb_score',
           s = 'budget2',
           data = movie_samp)

idx_min = movie_samp['imdb_score'].idxmin()
idx_max = movie_samp['imdb_score'].idxmax()

zippedidx = zip([idx_min, idx_max], [0.5,-0.5])

for idx, offset in zippedidx:
    year = movie_samp.loc[idx, 'title_year']
    score = movie_samp.loc[idx, 'imdb_score']
    title = movie_samp.loc[idx, 'movie_title']
    ax.annotate(xy = (year, score),
    xytext = (year + 1, score + offset),
    s = title + ' ({})'.format(score),
    ha = 'center',
    size = 16,
    arrowprops = dict(arrowstyle = 'fancy'))
ax.set_title('IMDB Score by Year',
             size = 25)
ax.grid(True)
```

```python
import pandas as pd
flights = pd.read_csv('data/flights.csv')
flights.head()
```

```python
flights['DELAYED'] = flights['ARR_DELAY'].ge(15).astype(int)
cols = ['DIVERTED',
        'CANCELLED',
        'DELAYED']
flights['ON_TIME'] = 1 - flights[cols].any(axis = 1)
cols.append('ON_TIME')
status = flights[cols].sum()
status
```

```python
flights.head()
```

```python
# checks whether any of the listed columns is 1
flights[['DIVERTED',
        'CANCELLED',
        'DELAYED']].any(axis = 1)

# if none of the columns are 1, add a 1 in a new column
flights['ON_TIME'] = 1 - flights[cols].any(axis = 1)
```

```python
import matplotlib.pyplot as plt
fig, ax_array = plt.subplots(2, 3, figsize = (18, 8))
(ax1, ax2, ax3), (ax4, ax5, ax6) = ax_array
fig.suptitle('2015 US Flights - Univariate Summary',
             size = 20)

ac = flights['AIRLINE'].value_counts()
ac.plot(kind = 'barh',
        ax = ax1,
        title = 'Airline')

oc = flights['ORG_AIR'].value_counts()
oc.plot(kind = 'bar',
        ax = ax2,
        rot = 0,
        title = 'Origin City')

dc = flights['DEST_AIR'].value_counts().head(10)
dc.plot(kind = 'bar',
        ax = ax3,
        rot = 0,
        title = 'Destination City')

status.plot(kind = 'bar',
            ax = ax4,
            log = True,
            title = 'Flight Status')

flights['DIST'].plot(kind = 'kde',
                     ax = ax5,
                     xlim = (0, 3000),
                     title = 'Distance KDE')

flights['ARR_DELAY'].plot(kind = 'hist',
                          ax = ax6,
                          title = 'Arrival Delay',
                          range = (0,200));
```

```python
hour = flights['SCHED_DEP'] // 100
minute = flights['SCHED_DEP'] % 100
df_date = flights[['MONTH', 'DAY']].assign(YEAR = 2015,
                                           HOUR = hour,
                                           MINUTE = minute)
df_date.head()
```

```python
flight_dep = pd.to_datetime(df_date)
flight_dep.head()
```

```python
flights.index = flight_dep
fc = flights.resample('W').size()
```

```python
fc.plot(figsize = (12,3),
        title = 'Flights per week',
        grid = True)
```

```python
fc_miss = fc.where(fc > 1000)
fc_miss.head()
```

```python
fc_intp = fc_miss.interpolate(limit_direction = 'both')
fc_intp.head()
```

```python
ax =fc_intp.plot(color = 'black',
                figsize = (16,4))
fc_intp[fc < 500].plot(linewidth = 10,
                       grid = True,
                       color ='0.8',
                       ax = ax)

ax.annotate('Missing Data',
            xy=(.8, .55), 
            xytext=(.8, .77), 
            xycoords='axes fraction', 
            ha='center', 
            size=20,
            arrowprops=dict())

ax.set_title('Flights per Week (Interpolated Missing Data)')
```

```python
flights.groupby('DEST_AIR')['DIST']\
        .agg(['mean', 'count'])\
        .query('count > 100')\
        .sort_values('mean')\
        .tail(10)\
        .plot(kind = 'barh',
              y = 'mean',
              rot = 0,
              legend = False,
              title = 'Average Distance per Destination');
        
```

```python
fs = flights.reset_index(drop=True)[['DIST', 'AIR_TIME']]\
            .query('DIST <= 2000').dropna()
fs.plot(x = 'DIST',
        y = 'AIR_TIME',
        kind = 'scatter',
        s = 1,
        figsize = (16,4));
```

```python
fs['DIST_GROUP'] = pd.cut(fs['DIST'],
                          bins = range(0, 2001, 250))
fs['DIST_GROUP'].value_counts().sort_index()
```

```python
normalize = lambda x: (x - x.mean()) / x.std()
```

```python
fs['TIME_SCORE'] = fs.groupby('DIST_GROUP')['AIR_TIME']\
                     .transform(normalize)
fs.head()
```

```python
ax = fs.boxplot(by = 'DIST_GROUP',
                column = 'TIME_SCORE',
                figsize = (16,4))
ax.set_title('Z-Scores for Distance Groups')
ax.figure.suptitle('');
```

```python
outliers = flights.iloc[fs[fs['TIME_SCORE'] > 6].index] 
```

```python
outliers = outliers[['AIRLINE',
                     'ORG_AIR',
                     'DEST_AIR',
                     'AIR_TIME',
                     'DIST',
                     'ARR_DELAY',
                     'DIVERTED']]
outliers.head(2)
```
